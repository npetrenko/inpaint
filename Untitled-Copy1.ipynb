```python
mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
with tf.device("//cpu:0"):
    train_dataset = tf.data.Dataset.from_tensor_slices((x_train, y_train))
    test_dataset = tf.data.Dataset.from_tensor_slices((x_test, y_test))

    true_data = train_dataset.concatenate(test_dataset)
```

```python
def preprocess(x, y):
    x = tf.cast(x, tf.float32)
    x -= 125.
    x /= 125.
    x = tf.expand_dims(x, axis=-1)
    
    #y = tf.one_hot(y, 10, axis=-1, dtype=tf.float32)
    return x#, y
```

```python
with tf.device("//cpu:0"):
    train_dataset = train_dataset.shuffle(x_train.shape[0]).batch(300).map(preprocess)
    test_dataset = test_dataset.batch(200).map(preprocess)

    handle = tf.placeholder(tf.string, [])
    iterator = tf.data.Iterator.from_string_handle(handle, train_dataset.output_types, train_dataset.output_shapes)

    train_iterator = train_dataset.make_initializable_iterator()
    test_iterator = test_dataset.make_initializable_iterator()
    
    next_elt = iterator.get_next()
```

```python
sess = tf.InteractiveSession()
```

```python
train_handle = sess.run(train_iterator.string_handle())
test_handle = sess.run(test_iterator.string_handle())
```

```python
next_elt
```

```python
sess.run(train_iterator.initializer)
sess.run(test_iterator.initializer)
```

```python
import tensorflow.keras.layers as layers
```

```python
decoder_model = tf.keras.models.Sequential([
    layers.Dense(256, activation="relu"),
    layers.Dense(128, activation="relu"),
    layers.Dense(128, activation="relu"),

    layers.Dense(28*28, activation=None),
    layers.Reshape([28,28,1])
])
```

```python
noise = tf.random_normal([10,6])
gened = decoder_model(noise)
```

```python
discriminator_model = tf.keras.models.Sequential([
    layers.Flatten(),
    layers.Dense(128, activation="relu"),
    layers.Dense(128, activation="relu"),
    layers.Dense(128, activation="relu"),

    layers.Dense(1, activation=None)
])
```

```python
decoder_output = decoder_model(tf.random_normal([300, 6]))
disc_false_preds = discriminator_model(decoder_output)

false_samples_logp = tf.log(tf.nn.sigmoid(disc_false_preds))
```

```python
false_samples_logp = tf.reduce_mean(false_samples_logp)
```

```python
dec_grads_gan = tf.gradients(false_samples_logp, decoder_model.variables)

labels = tf.ones(tf.shape(next_elt)[0])
labels = tf.concat([labels,
                    tf.zeros(300)
                   ], axis=0)

logits = tf.concat(
[
    discriminator_model(next_elt),
    disc_false_preds
], axis=0)

disc_grads = tf.gradients(
    -tf.nn.sigmoid_cross_entropy_with_logits(labels=labels, 
                                            logits=logits[...,0]), 
                                            discriminator_model.variables)
#encoder_grads = tf.gradients(-vae_loss, inference_model.variables)

dec_grads = dec_grads_gan
```

for grad in dec_grads_gan:

cum_vae = tf.sqrt(cum_vae)
cum_gan = tf.sqrt(cum_gan)

dec_grads = []

for gangrad, vaegrad in zip(dec_grads_gan, dec_grads_vae):
    dec_grads += [
        (0.9*gangrad / cum_gan + 0.1*vaegrad / cum_vae)/(1/cum_vae + 1/cum_gan)
    ]

```python
GAN_opt = tf.train.AdamOptimizer(learning_rate=0.0003)
```

```python
gan_decoder_train_op = GAN_opt.apply_gradients(zip(dec_grads, decoder_model.variables))
gan_disc_train_op = GAN_opt.apply_gradients(zip(disc_grads, discriminator_model.variables))
```

```python
while True:
    for epoch in range(3):
        print(epoch)
        try:
            while True:
                sess.run(gan_disc_train_op, {handle: train_handle})
        except tf.errors.OutOfRangeError:
            sess.run(train_iterator.initializer)
    
    for epoch in range(0):
        print(epoch)
        try:
            while True:
                sess.run(gan_decoder_train_op, {handle: train_handle})
        except tf.errors.OutOfRangeError:
            sess.run(train_iterator.initializer)
```
