```python
import tensorflow as tf
import numpy as np
import tensorflow.keras.backend as K
```

```python
learning_phase = K.learning_phase()
sess = tf.InteractiveSession()
K.set_session(sess)
```

```python
mnist = tf.keras.datasets.mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
with tf.device("//cpu:0"):
    train_dataset = tf.data.Dataset.from_tensor_slices((x_train, y_train))
    test_dataset = tf.data.Dataset.from_tensor_slices((x_test, y_test))

    true_data = train_dataset.concatenate(test_dataset)
```

```python
def preprocess(x, y):
    x = tf.cast(x, tf.float32)
    x -= 125.
    x /= 125.
    x = tf.expand_dims(x, axis=-1)
    x = tf.image.resize_images(x, (32,32))
    
    #y = tf.one_hot(y, 10, axis=-1, dtype=tf.float32)
    return x#, y
```

```python
with tf.device("//cpu:0"):
    train_dataset = train_dataset.shuffle(x_train.shape[0], reshuffle_each_iteration=False)\
    .shuffle(1000).batch(300).map(preprocess)
    
    test_dataset = test_dataset.shuffle(x_test.shape[0], reshuffle_each_iteration=False)\
    .batch(200).map(preprocess)

    handle = tf.placeholder(tf.string, [])
    iterator = tf.data.Iterator.from_string_handle(handle, 
                                                   train_dataset.output_types, 
                                                   train_dataset.output_shapes)

    train_iterator = train_dataset.make_initializable_iterator()
    test_iterator = test_dataset.make_initializable_iterator()
    
    next_elt = iterator.get_next()
```

```python
train_handle = sess.run(train_iterator.string_handle())
test_handle = sess.run(test_iterator.string_handle())
```

```python
next_elt
```

```python
sess.run(train_iterator.initializer)
sess.run(test_iterator.initializer)
```

```python
import tensorflow.keras.layers as layers
```

```python
inference_model = tf.keras.models.Sequential([
    layers.Flatten(),
    layers.Dense(512, activation="relu"),
    layers.Dense(256, activation="relu"),
    layers.Dense(24)
], name="inference_model")
```

```python
inference_output = inference_model(next_elt)
```

```python
inference_output
```

```python
inf_mu = inference_output[...,:12]
inf_logsigma = inference_output[...,12:]

inf_sample = tf.random_normal(tf.shape(inf_mu), 
                              inf_mu, 
                              tf.exp(inf_logsigma))
```

```python
inf_sample
```

```python
decoder_model = tf.keras.models.Sequential([
    layers.Dense(256),
#     layers.BatchNormalization(),
    layers.Activation("relu"),
    
    layers.Dense(512),
#     layers.BatchNormalization(),
    layers.Activation("relu"),
    
    layers.Dense(1024),
#     layers.BatchNormalization(),
    layers.Activation("relu"),
    
    layers.Dense(32*32),
    layers.Reshape([32,32,1])
])
```

```python
recovered = decoder_model(inf_sample)
recovered
```

```python
def build_vae_loss():
    rec_loss = tf.reduce_sum(tf.square(recovered - next_elt), axis=[1,2,3])

    kl_loss = -inf_logsigma + (tf.exp(inf_logsigma)**2 + inf_mu**2)/2
    kl_loss = tf.reduce_sum(kl_loss, axis=-1)

    return tf.reduce_mean(kl_loss + rec_loss)

vae_loss = build_vae_loss()
```

```python
discriminator_model = tf.keras.models.Sequential([
    layers.Flatten(),
    layers.Dense(512, activation="relu"),
    layers.Dense(256, activation="relu"),
    
    layers.Dense(1, activation=None)
])
```

```python
def build_gan_losses():
    num_samples = tf.shape(next_elt)[0]
    
    decoder_output = decoder_model(tf.random_normal([num_samples,12]))
    disc_false = discriminator_model(decoder_output)[...,0]

    false_samples_logp = tf.log(tf.nn.sigmoid(disc_false))
    dec_gain = tf.reduce_mean(false_samples_logp)
    
    labels = tf.concat([tf.ones(num_samples), 
                        tf.zeros(num_samples)], axis=0)

    disc_true = discriminator_model(next_elt)[...,0]
    
    #disc_gain = tf.reduce_mean(tf.log(tf.nn.sigmoid(disc_true)) + tf.log(1 - tf.nn.sigmoid(disc_false)))
    
    logits = tf.concat([
        disc_true,
        disc_false
    ], axis=0)
        
    disc_loss = tf.nn.sigmoid_cross_entropy_with_logits(labels=labels, logits=logits)
    disc_gain = -tf.reduce_mean(disc_loss)
    
        
    return dec_gain, disc_gain

dec_gain, disc_gain = build_gan_losses()
```

```python
def build_gan_update():
    opt = tf.train.AdamOptimizer(learning_rate=0.0002)

    decoder_upd = opt.minimize(-dec_gain, var_list=decoder_model.variables)
    encoder_upd = opt.minimize(vae_loss, var_list=inference_model.variables)
    disc_upd = opt.minimize(-disc_gain, var_list=discriminator_model.variables)
    
    return (decoder_upd, encoder_upd, disc_upd)
```

```python
def build_vae_update():
    opt = tf.train.AdamOptimizer(learning_rate=0.0002).minimize(vae_loss)
    return opt

vae_train_step = build_vae_update()
```

```python
decoder_upd, _, disc_upd = build_gan_update()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
from matplotlib import pyplot as plt
noise = np.random.normal(size=[6,12]).astype(np.float32)
ggg = decoder_model(tf.constant(noise))

def show_pics():
    arr = sess.run(ggg, {learning_phase: False})
    arr = np.concatenate([arr[i] for i in range(arr.shape[0])], axis=1)
    plt.imshow(arr[...,0])
    plt.show()
```

```python
for epoch in range(501):
    if epoch % 10 == 0:
        print(epoch)
        show_pics()
    
    try:
        while True:
            sess.run(vae_train_step, {handle: train_handle, learning_phase: True})
    except tf.errors.OutOfRangeError:
        sess.run(train_iterator.initializer)
```
